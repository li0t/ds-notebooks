```python
# importing dependencies

import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
%matplotlib inline
```

```python
# read data
recent_grads = pd.read_csv('data/recent-grads.csv')
```

```python
# display data info
recent_grads.info()
recent_grads.describe()
```

```python
# see first row
recent_grads.iloc[0]
```

```python
# see head
recent_grads.head()
```

```python
# see tail
recent_grads.tail()
```

```python
# remove empty rows
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]

print('Total rows were {0}, cleaned rows are {1}. {2} rows were deleted'.format(raw_data_count, cleaned_data_count, raw_data_count - cleaned_data_count ))
```

```python
# Sample_size vs Median scatter plot
sample_size_median_ax = recent_grads.plot(
    x="Sample_size", 
    y="Median", 
    kind="scatter", 
    title='Sample_size vs Median', 
    figsize=(7,7)
    )
```

```python
# Sample_size vs Unemployment_rate scatter plot
sample_size_unemployment_rate_ax = recent_grads.plot(
    x="Sample_size", 
    y="Unemployment_rate", 
    kind="scatter", 
    title='Sample_size vs Unemployment_rate', 
    figsize=(7,7)
    )
```

```python
# Full_time vs Median scatter plot
full_time_median_ax = recent_grads.plot(
    x="Full_time", 
    y="Median", 
    kind="scatter", 
    title='Full_time vs Median', 
    figsize=(7,7)
    )
```

```python
# ShareWomen vs Unemployment_rate scatter plot
sharewomen_unemployment_rate_ax = recent_grads.plot(
    x="ShareWomen", 
    y="Unemployment_rate", 
    kind="scatter", 
    title='ShareWomen vs Unemployment_rate', 
    figsize=(7,7)
    )
```

```python
# Men vs Median scatter plot
men_median_ax = recent_grads.plot(
    x="Men", 
    y="Median", 
    kind="scatter", 
    title='Men vs Median', 
    figsize=(7,7)
    )
```

```python
# Women vs Median scatter plot
women_median_ax = recent_grads.plot(
    x="Women", 
    y="Median", 
    kind="scatter", 
    title='Women vs Median', 
    figsize=(7,7)
    )
```
